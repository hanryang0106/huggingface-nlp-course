## Tokenizers
1. word-based tokenizers
each word gets assigned an ID, starting from 0 and going up to the size of the vocabulary. We need a custom token to present words that are not in our vocabulary. This is known as "unknown" token. One way to reduce the amount of unknown tokens is to go one level deeper, using a *character-based* tokenizer
2. character-based tokenizers
it splits the text into characters, rather than words. The vocabulary is much smaller, there are much fewer out-of-vocabulary tokens, since every word can be built from characters.
But this approach isn't perfect either. One could argue that each character doesn't mean a lot on its own, whereas that is the case with words. Another thing to consider is that we'll end up with a very large amount of tkens to be processed by our model
3. subword tokenizers
it relies on the principle that frequently used words should not be split into smaller subwords, but rare words should be decomposed into meaningful subwords. This approach is especially useful in agglutinative languages such as Turkish, Korean, where you can form arbitrarily long complex words by stringing together subwords.

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

sequence = "Using a Transformer network is simple"
tokens = tokenizer.tokenize(sequence)

print(tokens)
```

```python
tokenizer = AutoTokenizer.from_pretrained("albert-base-v1")
tokens = tokenizer.tokenize(sequence)

print(tokens)
# ALBERT tokenizers will add a long underscore in front of all the tokens that had a space bofore them, which is a convention used by sentencepiece tokenizers
```

```python
ids = tokenizer.convert_tokens_to_ids(tokens)

print(ids)
```

```python
decoded_string = tokenizer.decode([7993, 170, 13809, 23763, 2443, 1110, 3014])
print(decoded_string)
```
